# Financial Stock Data Analysis

### Reading Stock Data and using Pandas to calculate moving averages

```python
import numpy as np
import pandas as pd
from pandas_datareader import data as web
import pylab as pl
import matplotlib
% matplotlib inline
```

```python
msft = web.DataReader("MSFT", "google")
splk = web.DataReader("SPLK", "google")
```

```python
msft['Close'].plot(grid=True, figsize=(8, 5))
splk['Close'].plot(grid=True, figsize=(8, 5))
```

```python
pl.plot(msft['Close'])
pl.plot(splk['Close'])
pl.grid(True)
#pl.show()
```

```python
msft.describe()
```

```python
msft['11d']  = msft['Close'].rolling(11).mean()
msft['50d']  = msft['Close'].rolling(50).mean()
#msft[['Close', '11d', '50d']].head()
```

```python
pl.plot(msft['Close'])
pl.plot(msft['11d'])
pl.plot(msft['50d'])
pl.grid(True)
#pl.show()
```
